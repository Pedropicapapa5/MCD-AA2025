# Estadística descriptiva

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('/content/drive/MyDrive/Spotify Features 3.0.csv')
df
```

```python
df = df[['genre','artist_name','track_name','track_id','popularity','acousticness','danceability',
'duration_ms','energy','instrumentalness','key','liveness','loudness','mode',
'speechiness','tempo','valence','year_release']]
df.head()
```

```python
## Tipos de datos y descripción
df.info()
df.describe().T
```

- `popularity`, `danceability`, `energy`, `valence`, etc. son numéricas continuas.
- `genre`, `artist_name` y `track_name` son categóricas.

```python
## Comprobación de normalidad (paramétrico vs no paramétrico)
### Prueba de **Shapiro-Wilk** (muestra ≤ 5000) para cada variable numérica.
from scipy.stats import shapiro
variables_numericas = ['popularity','acousticness','danceability','duration_ms','energy',
'instrumentalness','liveness','loudness','speechiness','tempo','valence','year_release']
```

```python
normalidad = {}
for var in variables_numericas:
  stat, p = shapiro(df[var].sample(5000) if len(df) > 5000 else df[var])
  normalidad[var] = p
```

```python
normalidad_df = pd.DataFrame(list(normalidad.items()), columns=['Variable','p_value'])
normalidad_df['Distribución'] = np.where(normalidad_df['p_value']>0.05,'Paramétrica','No paramétrica')
normalidad_df
```

```python
## Estadísticos descriptivos
descriptivos = df[variables_numericas].describe().T
descriptivos[['mean','std','min','max']]
```

```python
## 5️⃣ Matriz de correlación y análisis
corr_matrix = df[variables_numericas].corr(method='pearson')
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Matriz de Correlación - Pearson')
plt.show()
```

- `energy` suele correlacionar positivamente con `loudness`.
- `acousticness` tiende a correlacionarse negativamente con `energy` y `loudness`.

```python
## Prueba de hipótesis
# ¿Existe correlación significativa entre `danceability` y `popularity`?**
H₀: ρ = 0 (no correlación)
H₁: ρ ≠ 0
```

```python
r, p_value = stats.pearsonr(df['danceability'], df['popularity'])
print(f"Coeficiente r = {r:.3f}, p-value = {p_value:.5f}")
# Debido que el p-value (0) fue menor que 0.05, se rechaza H₀ → hay relación significativa.
```

```python
## Visualizacion
# Dispersión con línea de regresión
sns.lmplot(x='danceability', y='popularity', data=df, line_kws={'color':'red'})
plt.title('Relación entre Danceability y Popularity')
plt.show()
```

```python
# Histogramas y boxplots
num_vars = ['popularity','energy','danceability','acousticness','valence']
for col in num_vars:
  plt.figure(figsize=(10,4))
  plt.subplot(1,2,1)
  sns.histplot(df[col], kde=True)
  plt.title(f'Histograma de {col}')


  plt.subplot(1,2,2)
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot de {col}')
  plt.show()
```

- Los resultados visuales y descriptivos permiten observar tendencias en cómo las características de audio se relacionan con la popularidad.
