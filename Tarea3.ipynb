# Estad√≠stica descriptiva

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('/content/drive/MyDrive/Spotify Features 3.0.csv')
df
```

```python
df = df[['genre','artist_name','track_name','track_id','popularity','acousticness','danceability',
'duration_ms','energy','instrumentalness','key','liveness','loudness','mode',
'speechiness','tempo','valence','year_release']]
df.head()
```

```python
## Tipos de datos y descripci√≥n
df.info()
df.describe().T
```

- `popularity`, `danceability`, `energy`, `valence`, etc. son num√©ricas continuas.
- `genre`, `artist_name` y `track_name` son categ√≥ricas.

```python
## Comprobaci√≥n de normalidad (param√©trico vs no param√©trico)
### Prueba de **Shapiro-Wilk** (muestra ‚â§ 5000) para cada variable num√©rica.
from scipy.stats import shapiro
variables_numericas = ['popularity','acousticness','danceability','duration_ms','energy',
'instrumentalness','liveness','loudness','speechiness','tempo','valence','year_release']
```

```python
normalidad = {}
for var in variables_numericas:
  stat, p = shapiro(df[var].sample(5000) if len(df) > 5000 else df[var])
  normalidad[var] = p
```

```python
normalidad_df = pd.DataFrame(list(normalidad.items()), columns=['Variable','p_value'])
normalidad_df['Distribuci√≥n'] = np.where(normalidad_df['p_value']>0.05,'Param√©trica','No param√©trica')
normalidad_df
```

```python
## Estad√≠sticos descriptivos
descriptivos = df[variables_numericas].describe().T
descriptivos[['mean','std','min','max']]
```

```python
## 5Ô∏è‚É£ Matriz de correlaci√≥n y an√°lisis
corr_matrix = df[variables_numericas].corr(method='pearson')
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Matriz de Correlaci√≥n - Pearson')
plt.show()
```

- `energy` suele correlacionar positivamente con `loudness`.
- `acousticness` tiende a correlacionarse negativamente con `energy` y `loudness`.

```python
## Prueba de hip√≥tesis
# ¬øExiste correlaci√≥n significativa entre `danceability` y `popularity`?**
H‚ÇÄ: œÅ = 0 (no correlaci√≥n)
H‚ÇÅ: œÅ ‚â† 0
```

```python
r, p_value = stats.pearsonr(df['danceability'], df['popularity'])
print(f"Coeficiente r = {r:.3f}, p-value = {p_value:.5f}")
# Debido que el p-value (0) fue menor que 0.05, se rechaza H‚ÇÄ ‚Üí hay relaci√≥n significativa.
```

```python
## Visualizacion
# Dispersi√≥n con l√≠nea de regresi√≥n
sns.lmplot(x='danceability', y='popularity', data=df, line_kws={'color':'red'})
plt.title('Relaci√≥n entre Danceability y Popularity')
plt.show()
```

```python
# Histogramas y boxplots
num_vars = ['popularity','energy','danceability','acousticness','valence']
for col in num_vars:
  plt.figure(figsize=(10,4))
  plt.subplot(1,2,1)
  sns.histplot(df[col], kde=True)
  plt.title(f'Histograma de {col}')


  plt.subplot(1,2,2)
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot de {col}')
  plt.show()
```

- Los resultados visuales y descriptivos permiten observar tendencias en c√≥mo las caracter√≠sticas de audio se relacionan con la popularidad.
