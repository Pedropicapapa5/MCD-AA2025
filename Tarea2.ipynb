# Música en SPOTIFY

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df_csv = pd.read_csv('/content/drive/MyDrive/SpotifyFeatures.csv')
df_csv
```

```python
type(df_csv)
```

```python
df = pd.DataFrame(df_csv)
```

```python
# Tipos de datos
print('\nTipos de datos:')
print(df.dtypes)
```

```python
# Asignar columna como índice
df = df.reset_index(inplace=False) # Reset index to ensure 'track_name' is a column
df = df.set_index('track_name')
print('\nDataFrame con índice = track_name:')
display(df.head())
```

```python
# Nueva columna con duración en minutos
df['duration_min'] = df['duration_ms']/60000
df.head()
```

```python
# Eliminar columna de duración en milisegundos
df = df.drop(columns=['duration_ms'])
df.head()
```

```python
print('\nDataFrame con nuevas columnas:')
display(df.head())
```

```python
# Filtrar registros
## Canciones con popularidad mayor a 90
filtro_populares = df[df['popularity'] > 90]
print('\nCanciones con popularidad mayor a 90:')
display(filtro_populares)
```

```python
# # Subconjuntos de datos
## Canciones de EXO
EXO_df = df[df['artist_name'] == 'EXO']
print('\nSubconjunto - Canciones de EXO:')
display(EXO_df)
```

```python
# Crear grupos de datos
## Agrupar por género y obtener promedio de popularidad
pop_por_genero = df.groupby('genre')['popularity'].mean()
print('\nPromedio de popularidad por género:')
display(pop_por_genero)
```
